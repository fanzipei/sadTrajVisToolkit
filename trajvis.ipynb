# Multiple User Visualization

There are two ways of visualizing multiple user on the same map. One is to construct a list, each element of which is the same format of single trajectory. Another way is using pandas dataframe, the dataframe has to contain four columns: uid, times, lat and lon. Note that we could also designate the colors, weights and icons of the trajectories to distinguish each user via passing a list of colors/weights/icons. Random colors could be generated by the new function random_color(size).

```python
visualize_traj(df, animation=True, period='PT60S', transition_time=100, color=None, duration='PT2H')
```
